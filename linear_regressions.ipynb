# **Linear Regression**

# **Importing the libraries**

```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# **Allocating Data**

```python
X = [[15], [18], [20], [22], [25], [28], [30], [32], [35], [38], [40], [42]]
Y = [20, 30, 50, 55, 70, 80, 95, 100, 130, 150, 170, 190]
print(X)
print(Y)
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y, color='red', linestyle='dashed', label='Predicted Line')
plt.xlabel("Temperature (Â°C)")
plt.ylabel("Ice Cream Sales")
plt.title("Temperature vs. Ice Cream Sales (Linear Regression)")
plt.legend()
plt.show()
```

# **Fit Data to the Model**

```python
r=LinearRegression()
r.fit(X,Y)
print(r.score(X,Y))
```

# **Prediction**

```python
print(r.predict([[32]]))
```

# **Fit Line**

```python
plt.scatter(X,Y)
plt.plot(X,r.predict(X),color='purple')
plt.show()
```
