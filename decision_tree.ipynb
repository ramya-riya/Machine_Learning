```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
```

```python
data=pd.read_csv("Ecommerce_analysis data.csv")
print(data.head())
```

```python
X = data[['Age', 'Frequency_of_Purchase', 'Product_Rating', 'Time_Spent_on_Product_Research(hours)']]
Y = data['Customer_Satisfaction']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=7)
print(X_test)
print(Y_test)
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train,Y_train)
```

```python
Y_pred = clf.predict(X_test)
print("Predicted values:\n", Y_pred)
print("Original Values: ",Y_test)
```

```python
print("Classification Report:\n", classification_report(Y_test, Y_pred,zero_division=1))
print("Accuracy Score: ", accuracy_score(Y_test, Y_pred))
```

```python
print("Actual vs Predicted:\t")
comparison = pd.DataFrame({'Actual': Y_test.values, 'Predicted': Y_pred})
print(comparison.head(10))
```
